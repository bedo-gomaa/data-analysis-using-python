**replace white space in each column to _**

```python
columns = ['index', 'Genre', 'Year', 'Movies Released', 'Gross', 'Tickets Sold',
           'Inflation-Adjusted Gross', 'Top Movie', 'Top Movie Gross (That Year)',
           'Top Movie Inflation-Adjusted Gross (That Year)']

for i, c in enumerate(columns):
    columns[i] = c.replace(" ", "_")

df.columns = columns
```

```python
df.head()
```

**what is the gross per year**

**every Genre and its Tickets_Sold**

```python
ticket_sold_per_gener =df.groupby('Genre')['Tickets_Sold'].mean()
ticket_sold_per_gener=ticket_sold_per_gener.sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have the 'ticket_sold_per_genre' Series
sns.barplot(x=ticket_sold_per_gener.index, y=ticket_sold_per_gener.values)
plt.xticks(rotation=90)

# Optional: Add labels and title to the plot
plt.xlabel('Genre')
plt.ylabel('Mean Tickets Sold')
plt.title('Mean Tickets Sold per Genre')

# Show the plot
plt.show()
```
